```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import wavfile
import scipy.fftpack as ft

%matplotlib inline
```

```python
def getValue(y, t):
    return y[int(44100.0 * t)]
def getRange(y, ti, tf):
    return y[int(44100.0 * ti):int(44100.0 * tf)]

rate, data = wavfile.read('note2.wav')  # reads wav
y = np.array([i[0] for i in data])      # takes the left channel
peak = np.max(y)
y = np.array([i / peak for i in y])
x = np.linspace(0, 167157.0/44100, 167157)
test = list(np.linspace(0, 3, 10))
samplex = getRange(x, 0.3, 0.5)
sampley = getRange(y, 0.3, 0.5)
plt.plot(samplex, sampley)
# y[200:300]
```

```python
yf = ft.fft(y)                  # applies fourier transform
plt.plot(np.abs(yf[0:4000]))    # takes positive values from graph
```

```python
yff = ft.ifft(yf)
plt.plot(yff)
```
